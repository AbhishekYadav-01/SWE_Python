```python
# Generating headers for the request
def gen_headers(sym):
    cookie = get_cookie(sym)
    hdrs = {
        'accept': '*/*',
        'accept-encoding': 'gzip',
        'accept-language': 'en-US,en;q=0.9',
        'cookie': cookie,
        'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Mobile Safari/537.36'
    }
    if sym == "NIFTY":
        hdrs['referer'] = 'https://www.nseindia.com/option-chain'
    elif sym == "HDFCBANK":
        hdrs['referer'] = 'https://www.nseindia.com/get-quotes/derivatives?symbol=HDFCBANK'
    return hdrs
```

```python
# Fetching and saving data
def fetch_data(url, hdrs, fname):
    try:
        res = requests.get(url=url, headers=hdrs)
        if res.status_code == 200:
            data = res.json()
            save_to_file(data, fname)
            print(f"Data is being saved to {fname}")
        else:
            print(f"Fetching data from {url} is failing, status: {res.status_code}")
    except Exception as err:
        print(f"Encountering an error: {str(err)}")
```

```python
# Saving data to file
def save_to_file(data, fname):
    with open(fname, 'w') as file:
        json.dump(data, file, indent=4)
```

```python
# Fetching NIFTY data
def fetch_nifty():
    url = "https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY"
    hdrs = gen_headers("NIFTY")
    fetch_data(url, hdrs, "NIFTY_option_chain.json")

# Fetching HDFCBANK data
def fetch_hdfc():
    url = "https://www.nseindia.com/api/option-chain-equities?symbol=HDFCBANK"
    hdrs = gen_headers("HDFCBANK")
    fetch_data(url, hdrs, "HDFCBANK_option_chain.json")
```

```python
# Running the main function
def run(test=False):
    if test:
        print("Running in testing mode...")
        fetch_nifty()
        fetch_hdfc()
        print("Testing is complete.")
    else:
        while True:
            fetch_nifty()
            fetch_hdfc()
            print("Data is being fetched. Waiting for 3 minutes...")
            time.sleep(180)

if __name__ == "__main__":
    run(test=True)  # Running with test=False for continuous fetching. For testing make test = true.
```

# Subtask 2

```python
import json
import pandas as pd
# Loading JSON data from a file
def load_json_data(filename):
    with open(filename, 'r') as file:
        return json.load(file)

```

```python
# Extracting the current price from JSON data
def extract_current_price(data):
    return data['records']['data'][0]['PE']['underlyingValue']
```

```python
# Extracting option chain data into a list of dictionaries
def extract_option_chain_data(option_chain):
    rows = []
    for item in option_chain:
        if 'CE' in item and 'PE' in item:
            rows.append({
                'strikePrice': item['strikePrice'],
                'expiryDate': item['expiryDate'],
                'call_open_interest': item['CE']['openInterest'],
                'put_open_interest': item['PE']['openInterest'],
                'call_ltp': item['CE']['lastPrice'],
                'put_ltp': item['PE']['lastPrice'],
                'call_change_in_oi': item['CE']['changeinOpenInterest'],
                'put_change_in_oi': item['PE']['changeinOpenInterest'],
                'call_volume': item['CE']['totalTradedVolume'],
                'put_volume': item['PE']['totalTradedVolume'],
            })
    return rows
```

```python
# Creating a DataFrame from option chain data
def create_dataframe(rows):
    return pd.DataFrame(rows)

```

```python
# Filtering data for the next weekly expiry
def filter_by_expiry(data, expiry_dates):
    next_weekly_expiry = expiry_dates[0]  # Assuming the first date is the next weekly expiry
    filtered_data = data[data['expiryDate'] == next_weekly_expiry]
    return filtered_data.sort_values(by='strikePrice')
```

```python

# Filtering rows above and below the current price
def filter_above_below_current(data, current_price):
    above = data[data['strikePrice'] > current_price].head(5)
    below = data[data['strikePrice'] < current_price].tail(5)
    return above, below

```

```python
# Displaying the results
def display_results(above, below, label):
    print(f"Rows above current price for {label}:")
    print(above)
    print(f"\nRows below current price for {label}:")
    print(below)
```

```python
# Handling NIFTY data
def process_nifty_data():
    nifty_data = load_json_data("NIFTY_option_chain.json")
    current_price_nifty = extract_current_price(nifty_data)
    nifty_option_chain = extract_option_chain_data(nifty_data['records']['data'])
    nifty_df = create_dataframe(nifty_option_chain)
    filtered_nifty_df = filter_by_expiry(nifty_df, nifty_data['records']['expiryDates'])
    above_current, below_current = filter_above_below_current(filtered_nifty_df, current_price_nifty)
    display_results(above_current, below_current, "NIFTY")
```

```python
# Handling HDFCBANK data
def process_hdfc_data():
    hdfc_data = load_json_data("HDFCBANK_option_chain.json")
    current_price_hdfc = extract_current_price(hdfc_data)
    hdfc_option_chain = extract_option_chain_data(hdfc_data['records']['data'])
    hdfc_df = create_dataframe(hdfc_option_chain)
    filtered_hdfc_df = filter_by_expiry(hdfc_df, hdfc_data['records']['expiryDates'])
    above_current, below_current = filter_above_below_current(filtered_hdfc_df, current_price_hdfc)
    display_results(above_current, below_current, "HDFCBANK")
```

```python
# Running the main execution function
def main():
    process_nifty_data()
    process_hdfc_data()

if __name__ == "__main__":
    main()
```

# Subtask 3

```python
# Loading and parsing JSON data
def load_data(file_path):
    with open(file_path, 'r') as f:
        return json.load(f)
```

```python
# Extracting option chain data and creating a list of rows
def extract_data(data):
    rows = []
    for itm in data['records']['data']:
        if 'CE' in itm and 'PE' in itm:
            rows.append({
                'strike': itm['strikePrice'],
                'expiry': itm['expiryDate'],
                'call_oi': itm['CE']['openInterest'],
                'put_oi': itm['PE']['openInterest'],
                'call_iv': itm['CE']['impliedVolatility'],
                'put_iv': itm['PE']['impliedVolatility'],
                'call_chg_oi': itm['CE']['changeinOpenInterest'],
                'put_chg_oi': itm['PE']['changeinOpenInterest'],
            })
    return rows
```

```python
# Creating a DataFrame from the extracted rows
def create_df(rows):
    return pd.DataFrame(rows)
```

```python
# Filtering DataFrame by the selected expiry date
def filter_expiry(df, expiry):
    return df[df['expiry'] == expiry]
```

```python
# Plotting Open Interest for calls and puts
def plot_oi(df, sym, expiry):
    plt.figure(figsize=(14, 7))
    plt.plot(df['strike'], df['call_oi'], label='Call OI', marker='o')
    plt.plot(df['strike'], df['put_oi'], label='Put OI', marker='o')
    plt.title(f'{sym} - Strike vs Open Interest for {expiry}')
    plt.xlabel('Strike Price')
    plt.ylabel('Open Interest')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plotting Implied Volatility and Change in Open Interest
def plot_iv_oi(df, sym, expiry):
    fig, ax1 = plt.subplots(figsize=(14, 7))
    ax1.set_xlabel('Strike Price')
    ax1.set_ylabel('Implied Volatility', color='tab:blue')
    ax1.plot(df['strike'], df['call_iv'], label='Call IV', color='tab:blue', marker='o')
    ax1.plot(df['strike'], df['put_iv'], label='Put IV', color='tab:cyan', marker='o')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Open Interest Change', color='tab:red')
    ax2.plot(df['strike'], df['call_chg_oi'], label='Call OI Chg', color='tab:red', marker='o')
    ax2.plot(df['strike'], df['put_chg_oi'], label='Put OI Chg', color='tab:orange', marker='o')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    fig.tight_layout()
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.title(f'{sym} - IV and OI Change for {expiry}')
    plt.grid(True)
    plt.show()
```

```python
# Processing and plotting data for the given file path and symbol
def main_task(file_path, sym):
    data = load_data(file_path)
    rows = extract_data(data)
    df = create_df(rows)
    expiry_dates = data['records']['expiryDates']
    selected_expiry = expiry_dates[0]  # Selecting the first expiry date
    filtered_df = filter_expiry(df, selected_expiry)

    # Plotting Open Interest and IV/OI Change
    plot_oi(filtered_df, sym, selected_expiry)
    plot_iv_oi(filtered_df, sym, selected_expiry)
```

```python
# Running the main tasks for NIFTY and HDFCBANK
if __name__ == "__main__":
    main_task("NIFTY_option_chain.json", "NIFTY")
    main_task("HDFCBANK_option_chain.json", "HDFCBANK")
```
